# Air Quality Data Set 
http://archive.ics.uci.edu/ml/datasets/air+quality

## Feature describton:
Missing values are tagged with -200 value. 

0. Date	(DD/MM/YYYY)

1. Time	(HH.MM.SS)

2. True hourly averaged concentration CO in mg/m^3 (reference analyzer) 

3. PT08.S1 (tin oxide) hourly averaged sensor response (nominally CO targeted)

4. True hourly averaged overall Non Metanic HydroCarbons concentration in microg/m^3 (reference analyzer)

5. True hourly averaged Benzene concentration in microg/m^3 (reference analyzer)

6. PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted)

7. True hourly averaged NOx concentration in ppb (reference analyzer)

8. PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted)

9. True hourly averaged NO2 concentration in microg/m^3 (reference analyzer)

10. PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted)

11. PT08.S5 (indium oxide) hourly averaged sensor response (nominally O3 targeted)

12. Temperature in Â°C

13. Relative Humidity (%)

14. AH Absolute Humidity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
data = pd.read_excel('AirQualityUCI.xlsx', parse_dates=[['Date','Time']])
Date = pd.read_excel('AirQualityUCI.xlsx')['Date']
data.head()
```

```python
data.describe()
```

```python
data.index = data['Date_Time']
data.pop('Date_Time')
data.head()
```

```python
data = data.fillna(0)
data = data[data != -200]
data.describe()
```

```python
%%time
means = pd.DataFrame()
for i in Date.astype(str).unique():
    means = pd.concat([means, data[i].mean()],axis=1)
means = means.T
for i in np.arange(0,means.shape[1]):
    means.iloc[:,i] = means.iloc[:,i].fillna(means.iloc[:,i].mean())
means.index = Date.astype(str).unique()
```

NMHC(GT) got too much nans

```python
NMHC_GT = means['NMHC(GT)']
means.pop('NMHC(GT)');
```

```python
noon = data['T'].at_time('12:00')
fig =plt.figure(figsize = (13,9))
n = fig.add_subplot(1,1,1)
n.plot(noon.index,noon,'p')
n.set_title('Temprature(time)')
n.set_xlabel('Date')
n.set_ylabel('Temprature')
midnigt =  data['T'].at_time('00:00')
n.plot(midnigt.index,midnigt,'p')
n.legend(['noon temp.','midnight temp.']);
```

```python
fig =plt.figure(figsize = (13,9))
n = fig.add_subplot(1,1,1)
n.plot(means.index,means['CO(GT)'])
n.set_title('Mean day temperature', fontsize = 15)
n.set_xlabel('Date')
n.set_ylim([-1,7])
n.set_xticks(means.index[np.arange(0,means.index.shape[0],40)])
n.set_ylabel('CO(GT)');
```

```python
plt.figure(figsize=(12,9))
plt.scatter(means['CO(GT)'], means['C6H6(GT)'], marker = 'o', alpha = .8,
            edgecolor='black',s = 100, c = means['T'],cmap = 'cividis')
plt.colorbar().set_label('temprature')
plt.xlabel('CO(GT)')
plt.ylabel('C6H6(GT)');
```

```python
first = means['CO(GT)']
second =  means['C6H6(GT)']

fig = plt.figure(figsize=(10,10))
pair = fig.add_subplot(2,2,1,xticklabels = [])
std_1 = fig.add_subplot(2,2,2, yticklabels = [])
std_2 = fig.add_subplot(2,2,3)
pair.scatter(first,second,edgecolor='black',color = '#FF9900', s = 60)
std_1.hist(second,orientation = 'horizontal',edgecolor='black',color = '#FF9900')
std_2.hist(first, edgecolor='black',color = '#FF9900')
std_2.invert_yaxis()
fig.subplots_adjust(wspace =0, hspace = 0);
```

```python
first = means['PT08.S1(CO)']
second =  means['NO2(GT)']

fig = plt.figure(figsize=(10,10))
h_2d = fig.add_subplot(2,2,1)
a= h_2d.hist2d(first, second, cmap = 'jet',bins = 12)
fig.colorbar(h_2d.contourf(a[0],cmap = 'jet'))
hex_2d = fig.add_subplot(2,2,2,sharex=h_2d, sharey=h_2d)
hex_2d.hexbin(first, second, cmap = 'jet',gridsize = 12);
```
